This notebook demonstrates how to use a **Chat Model** in LangChain.

Chat Models are designed for **conversational tasks**. They accept a list of structured messages (`SystemMessage`, `HumanMessage`, `AIMessage`) and return a message object — making them the preferred choice for chatbots, agents, and most modern LLM applications.

## Example — Querying a Chat Model

### Understanding the `temperature` parameter

`temperature` controls the **randomness** (creativity) of the model's output:

- **Low values (`0.0` – `0.3`)** — deterministic and focused. The model picks the most likely next token. Great for factual answers and structured output.
- **Medium values (`0.4` – `0.7`)** — balanced. A natural mix of accuracy and variety.
- **High values (`0.8` – `1.5`)** — creative and diverse. The model takes more risks, producing varied or surprising responses.

#### Recommended values by use case

| Use Case | Recommended Temperature | Why |
|----------|-------------------------|-----|
| **Factual Q&A / RAG** | `0.0` – `0.2` | Reduces hallucinations; gives consistent answers |
| **Code generation** | `0.0` – `0.3` | Code must be precise and deterministic |
| **Data extraction / classification** | `0.0` | Same input should always produce the same output |
| **Summarization** | `0.3` – `0.5` | Slight variation while staying faithful to the source |
| **Conversational chatbot** | `0.5` – `0.7` | Natural, engaging tone without going off-topic |
| **Creative writing / brainstorming** | `0.8` – `1.2` | Encourages novel ideas and diverse phrasing |
| **Poetry / story generation** | `1.0` – `1.5` | Maximizes creativity and variety |

> 💡 **Tip:** Start with `0.7` (a sensible default) and adjust based on whether the output feels too rigid or too random.

````python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI(model='gpt-4', temperature=0.9)
result = llm.invoke("What is the capital of India")
print(result)
print(result.content)
````

### What's happening above

1. `load_dotenv()` reads the `.env` file and loads `OPENAI_API_KEY` into the environment.
2. `ChatOpenAI(model='gpt-4', temperature=0.9)` creates a chat model client with high creativity.
3. `llm.invoke("...")` sends the input. LangChain auto-wraps the string into a `HumanMessage`.
4. The result is an **`AIMessage` object** — not a plain string. It contains:
   - `content` — the actual text response
   - `response_metadata` — token usage, model info, finish reason, etc.
   - `usage_metadata` — input/output token counts

### Key observations

- `print(result)` shows the **full message object** with all metadata.
- `print(result.content)` extracts just the **text** of the response.

### Key takeaway

Chat Models give you **structured, role-aware** interactions and rich metadata — ideal for chatbots, agents, RAG pipelines, and any multi-turn LLM application. For simple single-turn text completion, see [LLMs](./llm.ipynb).
