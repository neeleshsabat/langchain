# LLM — Text Completion Model

This notebook demonstrates how to use a **Large Language Model (LLM)** in LangChain.

LLMs are general-purpose text-completion models — they take a **string** as input and return a **string** as output. They have no built-in concept of conversational roles (system / user / assistant).

> 📖 **Reference notes:** [LangChain Components — Detailed](../NOTES/3.%20Langchain%20Components%20Detailed.md#llms-vs-chat-models)

## Setup

Make sure you have a `.env` file in the project root with your OpenAI API key:

```
OPENAI_API_KEY=sk-...
```

## Example — Querying an LLM

````python
from langchain_openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

llm = OpenAI(model='gpt-3.5-turbo-instruct')
result = llm.invoke("What is the capital of India")
print(result)
````

### What's happening above

1. `load_dotenv()` reads the `.env` file and loads `OPENAI_API_KEY` into the environment.
2. `OpenAI(model='gpt-3.5-turbo-instruct')` creates an LLM client. Note that `gpt-3.5-turbo-instruct` is a **completion-style** model — not a chat model.
3. `llm.invoke("...")` sends the text as input.
4. The result is a **plain string**, not a structured message object.

### Key takeaway

LLMs are best for **simple, single-turn text generation tasks** — summarization, completions, translations, etc. For conversational or multi-turn use cases, prefer [Chat Models](./chatmodels.ipynb).
